```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, confusion_matrix, classification_report)

# Dataset paths (update if your username/path differs)
train_dir = r"C:\Users\HP\ANN\data\train"
test_dir = r"C:\Users\HP\ANN\data\test"

# Quick check: list subfolders
print("Train classes:", os.listdir(train_dir))
print("Test classes:", os.listdir(test_dir))
```

```python
# Step 2: Load and preprocess images

IMG_SIZE = 64  # resize all images to 64x64

def load_images_from_folder(base_dir, img_size=IMG_SIZE):
    images = []
    labels = []
    class_names = sorted(os.listdir(base_dir))  # e.g. ['cats', 'dogs']
    
    for label, class_name in enumerate(class_names):
        class_path = os.path.join(base_dir, class_name)
        for fname in os.listdir(class_path):
            fpath = os.path.join(class_path, fname)
            img = cv2.imread(fpath)  # reads in BGR color by default
            if img is None:
                continue  # skip unreadable/corrupt files
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # convert to RGB
            img = cv2.resize(img, (img_size, img_size))
            images.append(img)
            labels.append(label)
    
    return np.array(images), np.array(labels), class_names

# Load training data
X_train, y_train, class_names = load_images_from_folder(train_dir)
print("Train data shape:", X_train.shape)
print("Train labels shape:", y_train.shape)
print("Classes:", class_names, "-> mapped to 0 and 1 respectively")

# Load test data
X_test, y_test, _ = load_images_from_folder(test_dir)
print("Test data shape:", X_test.shape)
print("Test labels shape:", y_test.shape)
```

```python
# Step 3a: Normalize pixel values to range [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

print("X_train range:", X_train.min(), "-", X_train.max())
print("X_test range:", X_test.min(), "-", X_test.max())

# Step 3b: Build the ANN architecture
model = Sequential([
    Flatten(input_shape=(IMG_SIZE, IMG_SIZE, 3)),   # Input layer: flattens 64x64x3 image into a 1D vector
    Dense(256, activation='relu'),                  # Hidden layer 1
    Dense(128, activation='relu'),                  # Hidden layer 2
    Dense(1, activation='sigmoid')                  # Output layer: binary classification (cat=0, dog=1)
])

model.summary()
```

```python
# Step 4: Compile the ANN

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("Model compiled successfully.")
```

```python
# Step 5: Train the ANN

history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 6: Evaluate the ANN on the test dataset

# Get predicted probabilities, then convert to class labels (0 or 1)
y_pred_probs = model.predict(X_test)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

# Calculate evaluation metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("=== ANN Test Set Performance ===")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1-Score : {f1:.4f}")
print()
print("Detailed classification report:")
print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
# Step 7: Confusion Matrix for the ANN

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix (raw counts):")
print(cm)

# Visualize it
plt.figure(figsize=(5, 4))
plt.imshow(cm, cmap='Blues')
plt.title('ANN Confusion Matrix')
plt.colorbar()

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Annotate each cell with its count
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                  ha="center", va="center",
                  color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```

```python
# Step 8: Train an SVM efficiently on the same preprocessed data

from sklearn.svm import SVC
from sklearn.decomposition import PCA
import time

# Flatten images into 1D vectors (same normalized pixel data the ANN used)
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

print("Flattened train shape:", X_train_flat.shape)
print("Flattened test shape:", X_test_flat.shape)

# Reduce dimensionality with PCA before SVM — this is the main speed-up
pca = PCA(n_components=100, random_state=42)
X_train_pca = pca.fit_transform(X_train_flat)
X_test_pca = pca.transform(X_test_flat)

print("PCA-reduced train shape:", X_train_pca.shape)
print(f"Variance retained: {pca.explained_variance_ratio_.sum()*100:.2f}%")

# Train SVM on the reduced features
start_time = time.time()

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_train_pca, y_train)

end_time = time.time()
print(f"SVM training completed in {end_time - start_time:.2f} seconds")
```

```python
# Step 9: Evaluate the SVM on the test set

y_pred_svm = svm_model.predict(X_test_pca)   # <-- using PCA-reduced test features

acc_svm = accuracy_score(y_test, y_pred_svm)
prec_svm = precision_score(y_test, y_pred_svm)
rec_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)

print("=== SVM Test Set Performance ===")
print(f"Accuracy : {acc_svm:.4f}")
print(f"Precision: {prec_svm:.4f}")
print(f"Recall   : {rec_svm:.4f}")
print(f"F1-Score : {f1_svm:.4f}")
print()
print("Detailed classification report:")
print(classification_report(y_test, y_pred_svm, target_names=class_names))

# Confusion matrix for SVM
cm_svm = confusion_matrix(y_test, y_pred_svm)
print("Confusion Matrix (SVM):")
print(cm_svm)

plt.figure(figsize=(5, 4))
plt.imshow(cm_svm, cmap='Oranges')
plt.title('SVM Confusion Matrix')
plt.colorbar()

tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

thresh = cm_svm.max() / 2.0
for i in range(cm_svm.shape[0]):
    for j in range(cm_svm.shape[1]):
        plt.text(j, i, format(cm_svm[i, j], 'd'),
                  ha="center", va="center",
                  color="white" if cm_svm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```
